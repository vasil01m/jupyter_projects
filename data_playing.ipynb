Such class loads a csv file and store it as a pandas dataframe. It is possible to provide a `clean=clean_dict` when instantiate an object. This `clean_dict` must be a python dictionary whose keys are columns name and values are possible undesidered columns values.

Then the dataframe will be created without those rows whose value in the given columns are in the `clean_dict`. In our case we don't want `phase` to be `1` or `5`.

```python
cleaning_dict = {
    'phase': [1, 5]
}
```

```python
data_obj = DataContainer(
    'data/flight_8066207_complete.csv',
    clean=cleaning_dict
)
```
